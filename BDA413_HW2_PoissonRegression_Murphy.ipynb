```python
for period in range(1,5):
    period==train['season']
    y = train["count"]
    x = train.drop(["count","datetime"])
    #z = train['windspeed'].loc[ii]
    plt.plot(x, y, 'o', color='rgbc'[period-1], label=str(period), alpha=0.5)
plt.xlabel("Baseline", size=17)
plt.ylabel("Follow-up", size=17)
#plt.zlabel("Windspeed", size=17)
plt.ylim(ymin=-0.5)
ha, lb = plt.gca().get_legend_handles_labels()
leg = plt.figlegend(ha, lb, "center right", numpoints=1, handletextpad=0.0001)
leg.draw_frame(False)
```

```python
poisson_mod = sm.Poisson(y, x)
poisson_res = poisson_mod.fit(method="newton")
print(poisson_res.summary())
```

```python
predVals = poisson_res.predict(y)
```

```python
plt.plot(range(len(y)), y, 'r*-', range(len(y)), predVals,  'bo-', alpha = 0.7)
plt.plot(range(len(x)), x, 'r*-', range(len(x)), predVals, 'bo-', alpha= 1)
plt.title('Train dataset Real vs. Predicted Values')
plt.legend(['Real Values', 'Predicted Values'])
plt.show()
```
